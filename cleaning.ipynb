```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('startup_funding.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Remarks'],inplace=True)
df.set_index('Sr No',inplace=True)
df.rename(columns={
    'Date dd/mm/yyyy':'date',
    'Startup Name':'startup',
    'Industry Vertical':'vertical',
    'SubVertical':'subvertical',
    'City  Location':'city',
    'Investors Name':'investors',
    'InvestmentnType':'round',
    'Amount in USD':'amount (Cr)'    
    
},inplace=True)

df.head()
```

```python
df['amount (Cr)'].isnull().sum()
```

```python
df['amount (Cr)'] = df['amount (Cr)'].fillna('0')
df.info()
```

```python
df['amount (Cr)'] = df['amount (Cr)'].fillna('0')
df['amount (Cr)'] = df['amount (Cr)'].str.replace(',','').replace('undisclosed','0')
df['amount (Cr)'] = df['amount (Cr)'].str.replace('unknown','0').replace('Undisclosed',0)
df['amount (Cr)'] = pd.to_numeric(df['amount (Cr)'], errors='coerce')
```

```python
df['amount (Cr)'] = df['amount (Cr)'].astype('float')
```

```python
df['amount (Cr)'].isna().sum()
```

```python
df['amount (Cr)'] = df['amount (Cr)'].fillna(0)
df['amount (Cr)'].isnull().sum()
```

```python
df.info()
```

```python
def to_cr(usd):
    inr_cr = usd * 82.5
    return inr_cr/10000000
df['amount (Cr)'] = df['amount (Cr)'].apply(to_cr)
df.head()
```

```python
df['date'] = df['date'].str.replace('05/072018','05/07/2018')
df['date'] = pd.to_datetime(df['date'],format="%d/%m/%Y",errors='coerce')
df.head()
```

```python
df[df['date'].isna()]
df.dropna(subset=['date','city'],inplace=True)
df.info()
```

```python
df['investors'].fillna("unknown",inplace=True)
df['investors'].isnull().sum()
```

```python
investors_list = sorted(set(df['investors'].str.split(',').sum()))
investors_list = [i.strip().lower() for i in investors_list]
investors_list
```

```python
for i in investors_list:
    if '&' in i:
        investors_list.remove(i)
        investors_list.extend(i.split(' & '))
    elif 'and' in i:
        investors_list.remove(i)
        investors_list.extend(i.split(' and '))
investors_list=sorted(set(investors_list))
print(investors_list)
len(investors_list)
```

```python
def unify_investors_col(row):
    if len(str(row['investors']).split(','))>1:
        temp = str(row['investors']).split(',')
        for i in temp:
            if '&' in i:
                temp.remove(i)
                temp.extend(i.split(' & '))
            elif 'and' in i:
                temp.remove(i)
                temp.extend(i.split(' and '))

        for i in range(len(temp)):
            temp[i] = temp[i].strip().lower()
            if (len(temp[i].split(' '))>1):
                name = temp[i].split(' ')[:-1]
                if " ".join(name) in investors_list:
                    temp[i] = " ".join(name)
        return ",".join(temp)
    else:
        return str(row['investors']).lower()
    
df['investors'] = df.apply(unify_investors_col, axis = 1)
len(sorted(set(df.investors.str.split(',').sum())))
            
```

```python
df.insert(2,'t_name',df['startup'].str.lower())
s=sorted(df.t_name.unique())
print(len(s))
s
```

```python
def unify_startup_col(row):
    temp = str(row['t_name']).split(' ')

    if len(temp)>1:
        if "".join(temp) in s:
            return "".join(temp)
        else:
            return row['t_name']
    else:
        return row['t_name']

df['startup']=df.apply(unify_startup_col,axis=1)
df.drop(columns=['t_name'],inplace=True)
len(df.startup.unique())
```

```python
df.startup = df.startup.str.capitalize()
print(sorted(df.startup.unique()))
```

```python
df['startup'] = df['startup'].str.replace(r'\\\\x[a-fA-F0-9]{2}',"",regex=True)
df.dropna(subset=['investors'],inplace=True)
df['investors'] = df['investors'].str.replace(r'\\\\x[a-fA-F0-9]{2}',"",regex=True)
def fix_byju(name):
    if 'byju' in name or 'Byju' in name:
        return 'Byjus'
    else:
        return name
df['startup'] = df['startup'].apply(fix_byju)
print(sorted(df.startup.unique()))
```
